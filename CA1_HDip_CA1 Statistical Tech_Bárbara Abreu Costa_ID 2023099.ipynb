### 2.2 - The number of industrial injuries on average per working week in a factory is 0.75. 

Assuming that the distribution of injuries follows a Poisson distribution, find the probability that in a particular week there will be no more than two accidents. Stating the answers alone here is not sufficient - you should show clearly how you found your answer, justify any decisions made, and include plots where appropriate.

Where:

    mu = 0.75 is the mean number of accidents
   
We create a Poisson distribution object with mean mu, where: 
    
    dist = poisson(mu)7

```python
from scipy.stats import poisson

dist = poisson(0.75)
prob = dist.cdf(2)

print("Probability of no more than two accidents:", prob)
```

## Task 3 - Probability (Continuous):

The time a person spends at Dublin Zoo is Normally distributed with a mean of 90 minutes and a
standard deviation of 10 minutes.
Using this distribution, answer the following:

â If a visitor is selected at random, find the probability that they will spend at most 85 minutes visiting
the zoo.

â If a visitor is selected at random, find the probability that they will spend at least 100 minutes
visiting the zoo.

â Given that you know that a particular visitor has spent longer than average visiting the Zoo, what is
the probability that they have spent more than 100 minutes there?

Stating the answers alone here is not sufficient - you should show clearly how you found your answer, justify
any decisions made, and include plots where appropriate.

#### First we'll need to import a library

```python
from scipy.stats import norm
```

Where:

    loc = is the mean number of accidents, 90 minutes
    scale = is standard deviation, 10 minutes

#### â The probability of a visitor spending *at most* 85 minutes:

```python
prob85 = norm.cdf(85, loc=90, scale=10)
print("Probability of spending at most 85 minutes:", prob85)
```

So the probability of a visitor spending at most 85 minutes at the zoo is about 0.308 or 30.85%.

```python
mu = 90
sigma = 10

# Define the x-axis range for the plot
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)

# Calculate the corresponding y-axis values (pdf)
y = norm.pdf(x, loc=mu, scale=sigma)

# Create the plot
fig, ax = plt.subplots(figsize = (8,5))
ax.plot(x, y, color='red', linewidth = 2)

# Shade the area corresponding to the probability of spending at most 85 minutes
x_fill = np.linspace(mu - 4*sigma, 85, 100)
y_fill = norm.pdf(x_fill, loc=mu, scale=sigma)
ax.fill_between(x_fill, y_fill, color='pink')

# Add labels and title to the plot
ax.set_xlabel('Time in Minutes')
ax.set_ylabel('Probability Density')
ax.set_title('Standard Normal Distribution')

plt.show()
```

Above we can see a plot that shows the normal distribution

#### â The probability of a visitor spending *at least* 100 minutes:

```python
prob100 = 1 - norm.cdf(100, loc=90, scale=10)
print("Probability of spending at most 85 minutes:", prob100)
```

It means there is a probability of about 0.159 or 15.9% that a randomly selected visitor to the zoo will spend at least 100 minutes there.

```python
mu = 90
sigma = 10

# Define the x-axis range for the plot
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)

# Calculate the corresponding y-axis values (pdf)
y = norm.pdf(x, loc=mu, scale=sigma)

# Create the plot
fig, ax = plt.subplots(figsize=(8,5))
ax.plot(x, y, color='red', linewidth=2)

# Shade the area corresponding to the probability of spending more than 100 minutes
x_fill = np.linspace(100, mu + 4*sigma, 100)
y_fill = norm.pdf(x_fill, loc=mu, scale=sigma)
ax.fill_between(x_fill, y_fill, color='pink')

# Add labels and title to the plot
ax.set_xlabel('Time in Minutes')
ax.set_ylabel('Probability Density')
ax.set_title('Standard Normal Distribution')

plt.show()
```

Above we can see a plot that shows the normal distribution

#### â The probability of a visitor spending *more than* 100 minutes

Here, we'll need to use the conditional probability formula:

       P(A|B) = P(A and B) / P(B)
    
    Where:
        A = is the event that the visitor spends more than 100 minutes at the Zoo
        B = is the event that the visitor has spent longer than average visiting the Zoo

As was given, the mean time spent at the zoo is 90 minutes, so we'll assume:
      
        "Longer than average" means more than 90 minutes
        
        

```python
p_a = 1 - norm.cdf(100, loc=90, scale=10)
```

```python
p_b = 1 - norm.cdf(90, loc=90, scale=10)
```

```python
p_a_given_b = p_a / p_b
print(p_a_given_b)
```

There is a probability of spending more than 100 minutes given longer than average: 0.31731050786291415 or approximately 32%

```python
from scipy.stats import norm

# mean and standard deviation of the distribution
mean = 90
std_dev = 10

# probability of spending longer than average time at the zoo
prob_longer_than_avg = 0.5

# probability of spending more than 100 minutes given that they spent longer than average
prob_more_than_100_given_longer_than_avg = (1 - norm.cdf(100, loc=mean, scale=std_dev)) / prob_longer_than_avg

print("Probability of spending more than 100 minutes given longer than average:", prob_more_than_100_given_longer_than_avg)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu = 90
sigma = 10

# Define the x-axis range for the plot
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)

# Calculate the corresponding y-axis values (pdf)
y = norm.pdf(x, loc=mu, scale=sigma)

# Create the plot
fig, ax = plt.subplots(figsize=(8,5))
ax.plot(x, y, color='red', linewidth=2)

# Shade the area corresponding to the probability of spending more than 100 minutes
x_fill = np.linspace(100, mu + 4*sigma, 100)
y_fill = norm.pdf(x_fill, loc=mu, scale=sigma)
ax.fill_between(x_fill, y_fill, color='purple')

# Shade the area corresponding to the probability of spending longer than average
x_fill2 = np.linspace(mu, mu + 4*sigma, 100)
y_fill2 = norm.pdf(x_fill2, loc=mu, scale=sigma)
ax.fill_between(x_fill2, y_fill2, color='pink')

# Shade the area corresponding to the probability of spending more than 100 minutes given longer than average
x_fill3 = np.linspace(100, mu + 4*sigma, 100)
y_fill3 = norm.pdf(x_fill3, loc=mu, scale=sigma) * (y_fill2 / y_fill)
ax.fill_between(x_fill3, y_fill3, color='purple')

# Add labels and title to the plot
ax.set_xlabel('Minutes')
ax.set_ylabel('Probability Density')
ax.set_title('Normal Distribution')

plt.show()
```

In the plot above shaded area in purple represents the conditional probability of spending more than 100 minutes given longer than average.
